```python
tips = pd.read_csv("/content/drive/MyDrive/tips.csv")


sns.barplot('sex','total_bill', data=tips)

plt.xlabel('sex')
plt.ylabel('total_bill')
plt.legend()
plt.title('bar plot')
plt.show()
```

```python
tips = pd.read_csv("/content/drive/MyDrive/tips.csv")
colors=[0.0,0.2,0.3,0.4]
plt.scatter('day', 'total_bill', data=tips, marker='*',cmap="rainbow", s=25)
plt.xlabel('day')
plt.ylabel('total_bill')
plt.title('scatter plot')
plt.colorbar()

plt.ylim(6,12)
plt.show()
```

```python
tips = pd.read_csv("/content/drive/MyDrive/tips.csv")
sns.jointplot('smoker', 'tip', data=tips)
sns.set_style('dark')
sns.set_context('paper', font_scale=2)

plt.xlabel('smoker')
plt.ylabel('tip')
plt.title('joint plot')
plt.show()

```

```python
sns.pairplot(tips,hue='smoker')

```

```python
sns.countplot(x='state', data=data)
plt.xticks(fontsize=1.5)
```

```python
sns.boxplot(x='day', y='tip', data=tips, hue='sex')
plt.legend(loc=0, fontsize=15)
```

```python
sns.violinplot(x='sex', y='total_bill', hue='day', data=tips)
plt.legend(fontsize=10)
```

```python
sns.stripplot(x='day', y='tip', hue='smoker', data=tips, jitter=True, dodge=True, palette='magma')
sns.set_style('white')
plt.legend(loc=0)
```

```python
sns.violinplot(x='day', y='tip', data=tips)
sns.swarmplot(x='day', y='tip', data=tips, color='w')
plt.title('swarn plot', c='c', fontsize=30)
```

```python
import squarify 
data = pd.read_csv("/content/drive/MyDrive/schools_detailed.csv.zip")
data_df=data.groupby('state').size()
labels=data_df.index
sizes=data_df.values
colors=[plt.cm.tab10(i/float(len(labels))) for i in range(len(labels))]
plt.figure(figsize=(20,30), dpi=100)
squarify.plot(sizes=sizes, label=labels, color=colors, alpha=1)
plt.axis('off')
```
