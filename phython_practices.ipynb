```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
import seaborn as sns
```

```python
x=np.arange(0,100,5)
y1=x*2
y2=x*3

plt.plot(x,y1,lw='3',c='g',marker='*',label='multiple of 2')
plt.plot(x,y2,lw='3',c='c',marker='d',label='multiple of 3')

print(x,y1,y2)
f1={'family':'cambria','size':30,'color':'c'}
f2={'family':'cambria','size':15,'color':'g'}
plt.xlabel("y axis",fontdict=f2)
plt.ylabel("x axis",fontdict=f2)
plt.title("line plot",fontdict=f1)
plt.grid(c='r',ls='dotted',lw="2")


plt.legend()
plt.show()
```

```python
x=np.arange(0,5,1)
y1=x*2
y2=x*3
print(x,y1,y2)
plt.plot(x,y1,x,y2)
plt.legend(['multiple of 2','multiple of 3'],loc="upper right",framealpha=2,fancybox=True,shadow=False,facecolor='green',edgecolor='black')
plt.show()
```

```python
#scatter plot

x=[1,2,3,4,5,6,7,8,9]
y1=[10,20,30,40,50,60,70,80,90]
y2=[10,20,30,40,50,60,70,80,90]
random.shuffle(y1)
random.shuffle(y2)
print(x,y1,y2)

plt.scatter(x,y1,label='shuffle 1',color='g',marker='*',s=30,alpha=1)

plt.scatter(x,y2,label='shuffle 2',color='c',marker='s',s=35,alpha=0.5)
plt.legend()
plt.show()
```

```python
x=['a','b','c','d','e']
y=[10,20,30,25,13]
plt.xlabel('alphabets')
plt.ylabel('numbers')
plt.title('barchart')
c=['b','r','c','g','pink']

plt.bar(x,y,color=c,width=0.5)

plt.show()
```

```python
year=[2021,2022,2023,2024,2025]
placement=[568,458,425,256,563]

cse=[75,65,45,52,95]
it=[56,65,156,254,56]
ece=[85,263,45,89,75]
eee=[54,65,25,65,78]

ece_start=[cse[i]+it[i] for i in range(len(cse))]
eee_start=[ece_start[i]+ece[i] for i in range(len(ece))]

plt.bar(year,cse,color='pink',label='cse')
plt.bar(year,it,bottom=cse,color='g',label='it')
plt.bar(year,ece,bottom=ece_start,color='c',label='ece')
plt.bar(year,eee,bottom=eee_start,color='yellow',label='eee')
plt.legend()

plt.show()
```
